# Blocking Strategies

## Objective
Reduce the number of comparisons by grouping similar records.

### Simple Blocking

```python
df["block"] = df["last_name"].str[0]
df[["last_name", "block"]].head()
```

### Multi-blocking

Compensates for missing/incorrect fields
<br>Improves recall

```python
df["block1"] = df["last_name"].str[:2]
df["block2"] = df["first_name"].str[:1]
df["block3"] = df["dob"].str[:4]
```

Multi-blocking improves recall by compensating for missing or incorrect fields.
