# Blocking Strategies

## Objective
Reduce the number of comparisons by grouping similar records.

```python
import pandas as pd

df_A = pd.read_pickle("../data/processed/df_A_encoded.pkl")
df_B = pd.read_pickle("../data/processed/df_B_encoded.pkl")
```

### Multi-blocking

Compensates for missing/incorrect fields
<br>Improves recall

```python
def create_blocks(df):
    df["block1"] = df["last_name"].str[:2]
    df["block2"] = df["first_name"].str[:1]
    df["block3"] = df["dob"].str[:4]
    return df

df_A = create_blocks(df_A)
df_B = create_blocks(df_B)
```
